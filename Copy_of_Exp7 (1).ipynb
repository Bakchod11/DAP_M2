```python
# TASK 3
import pandas as panda
import numpy as np

data = panda.read_csv('Share_data.csv')

data = data.to_numpy()
data

# MEAN
print(data)
opene = data[:,1:2]
print(opene.mean())
closee = data[:,2:]
print(closee.mean())


#MODE
from scipy import stats

a = stats.mode(opene)
b = stats.mode(closee)
print(a)
print(b)

#MEDIAN

print(np.median(opene))
print(np.median(closee))

# STD
print(np.std(opene))
print(np.std(closee))

# RMSE (GOOD LUCK REMEMEBRING IT)
print(np.sqrt(np.mean((opene-closee)**2)))
```
