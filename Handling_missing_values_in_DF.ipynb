# Option 1: Filling null values with fillna() in dataframe


```python
from google.colab import files
upload = files.upload()
import io
import pandas as pd
df_table = pd.read_table(io.BytesIO(upload['Student_Data_2.txt']))
df_table
```

```python
df = df_table.isnull() # it returns a table of boolean values: "False" if value is correct and "True" if value is NaN
print(df)
```

```python
df1 = df_table.fillna(value = 10) # it fills the table with zero (specified value) in place of True 
print(df1)
```

```python
df = df1.isnull() # it returns a table of boolean values: "False" if value is correct and "True" if value is NaN
print(df)

df1.isnull().sum().sum()
```

```python
# Filling null value with previous value in the same column
df1 = df_table.fillna(method = 'pad')  # or df1 = df_table.fillna(method = 'ffill')
print(df1)
```

```python
#Filling null value with next value in the same column
df2 = df_table.fillna(method = 'bfill') 
print(df2)
```

```python
# filling null value with value from the earlier column and same row
df3 = df_table.fillna(method = 'pad', axis = 1) 
print(df3)
```

```python
# filling null value with value from the next column and same row
df4 = df_table.fillna(method = 'bfill', axis = 1) 
print(df4)
```

```python
# filling specific values in different columns - Need a dictionary
df5 = df_table.fillna({"DAP" : df_table["DAP"].mean(), "PS" : 18, "IEM":10, "WT": 18})
print(df5)
```

```python
# Filling all missing values with statstical values of the specified column - mean()
df7 = df_table.fillna(value = df_table["DAP"].mean())
print(df7)
```

```python
df8 = df_table.fillna(value = df_table["DAP"].max())
print(df8)
```

```python
df9 = df_table.fillna(value = df_table["DAP"].min())
print(df9)
```

```python
# Replacing all the same specific values of dataframe by specified value used as parameter
#df10 = df_table.replace(to_replace = 14, value = 20, inplace = True)
#print(df10)
#df_table
import numpy as np
df12 = df_table.replace(to_replace = np.NaN, value = 20, inplace = True)
df_table
```

```python
# replacing null values with numpy 
import numpy as np
df11 = df_table.replace(to_replace = np.NaN, value = 20)
print(df11)
```

```python
fn = "/content/gdrive/My Drive/Data Analysis with Python/faculty_data2.txt"
#fn1 = "/content/gdrive/My Drive/Data Analysis with Python/faculty_data1.txt"
```

```python
# reading a file with read() function
file1 = open(fn,"r")
data = file1.read()
print(data)
```

```python
# converting text file to dictionary as per data records.
file1 = open(fn,"r")
fd2 = file1.readlines()
print(fd2)
i=0
ind = 0
faculty_dict = {}
while(i<6): 
  faculty_list = []
  for j in range(i, len(fd2), 6):
    faculty_list.append(fd2[j].strip())
  faculty_dict.update({ind : faculty_list})
  i+=1
  ind+=1
    #faculty_data.update({i : })

print(faculty_dict)
```

```python
import pandas as pd1
faculty_data = pd1.DataFrame(faculty_dict)
faculty_data 
```

```python
faculty_data_new = faculty_data.rename(columns = {0 :"Name", 1:"Designation", 2:"Department", 3:"Education",4:"YoE", 5: "Address"})
faculty_data_new
```

#Replacing values in the dataframe with desired values

```python

import numpy as np
faculty_data_new['Designation'] = faculty_data_new['Designation'].replace("", np.nan)
faculty_data_new
faculty_data_new['YoE'] = faculty_data_new['YoE'].replace("", np.nan)
faculty_data_new
faculty_data_new['Address'] = faculty_data_new['Address'].replace("", "Borivali")
faculty_data_new
faculty_data_new['Department'] = faculty_data_new['Department'].replace("CE", "Computer Engineering")
faculty_data_new
```

```python
# making entire row as NaN for the spacified missing value from the column of dataframe 
faculty_data_new.loc[faculty_data_new['Department'] == ""] = np.nan   
faculty_data_new
```
