# Adding and deleting columns and rows to existing dataframe

```python
# adding 
import pandas as pd
marks = {'Rollno' : ['A1','A2','A3'], 'DAP': [12,14,16], 'PS' : [14,17,11]}
marks

df_marks = pd.DataFrame(marks)
df_marks
print(df_marks)
print("Number of records in dataframe :" + str(len(df_marks)))
```

```python
# Adding row through dictionary - first create a dictionary and use "append function"

df_row = {'Rollno': 'A4', 'DAP': 19, 'PS': 14}
df1 = df_marks.append(df_row, ignore_index = True)
df1

#df2 = df_marks.append(df_row, ignore_index = False) # it can append disctionary only if ignore_index is "True"
#df2
```

```python
# Adding new colums to the dataframe - creating dataframe

marks = {'Rollno' : ['A1','A2','A3'], 'DAP': [12,14,16], 'PS' : [14,17,11]}
marks

df_marks = pd.DataFrame(marks)
df_marks
```

```python
# Adding column through list in a DataFrame - "AWT" marks

AWT = [14,17,10]
AWT

df_marks['AWT'] = AWT   
df_marks

#df_marks['AWT'] = [20,18,20]
#df_marks

#df_marks['AWT'] = [20,18] # it gives an error - size of the list should be same as number of rows in the dataframe
#df_marks
```

```python
# creating dataframe through user input - using append function

df = pd.DataFrame(columns=['AWT'])
df

for i in range(len(df_marks)):
  m = int(input("Enter Marks"))
  df = df.append({'AWT': m}, ignore_index = True)
df
```

```python
# Adding columns from different dataframe to the existing dataframe - using concat function

New_marks = pd.concat([df_marks,df], axis=1)
New_marks
```

```python
# Adding rows index wise along with columns from different dataframe to the existing dataframe - when axis = 0

New_marks = pd.concat([df_marks,df], axis=0)
New_marks
```

```python
# Shorter way to create dataframe using concat function.

new = pd.concat([pd.DataFrame([i], columns=['AWT']) for i in range(5)],ignore_index=True)
new
```

```python
# another way to create dataframe - using list

df = pd.DataFrame([[11, 12], [13, 14]], columns=list('AB'), index=[0, 1])
df
```
