```python
# conversion of file ito dataframe using readlines() function
file1 = open(fn,"r")
fd2 = file1.readlines()
print(fd2)
i=0
f1 = []
while(i<4): 
  f1.append(fd2[i].strip())  # res.append(re.sub('\n', '', sub)) string = lis.strip()
  i+=1  
print(f1)


f2 = []
for i in range(4, 8, 1):
  f2.append(fd2[i].strip())

print(f2)

f3 = []
for i in range(8, 12, 1):
  f3.append(fd2[i].strip())

print(f3)

f4 = []
for i in range(12, 16, 1):
  f4.append(fd2[i].strip())

print(f4)

f5 = []
for i in range(16, 20, 1):
  f5.append(fd2[i].strip())

print(f5)

f6 = []
for i in range(20, len(fd2), 1):
  f6.append(fd2[i].strip())

print(f6)
```

```python
# converting text file to dictionary as per data records.
file1 = open(fn,"r")
fd2 = file1.readlines()
print(fd2)
i=0
ind = 0
faculty_dict = {}
while(i<4): 
  faculty_list = []
  for j in range(i, len(fd2), 4):
    faculty_list.append(fd2[j].strip())
  faculty_dict.update({ind : faculty_list})
  i+=1
  ind+=1
    #faculty_data.update({i : })

print(faculty_dict)
```

```python
# conveting dictionay to dataframe
import pandas as pd1
faculty_data = pd1.DataFrame(faculty_dict)
print(faculty_data)
print(type(faculty_data))

r,c = faculty_data.shape
print(r)
print(c)
```

```python
# renaming of the columns of teh dataframe
faculty_data_new = faculty_data.rename(columns = {0 :"NAME", 1:"DESIGNATION", 2:"DEPARTMNET", 3:"EDUCATION"})
print(faculty_data_new)
```

```python
#Changing the name of every column of dataframe as per requirement
r,c = faculty_data.shape
for i in range(0 , c, 1):
  col_name = input("Enter column name" + str(i))
  faculty_data = faculty_data.rename(columns = {i:col_name})
print(faculty_data)
```
